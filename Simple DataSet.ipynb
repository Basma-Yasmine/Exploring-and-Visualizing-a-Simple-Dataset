```python
plt.figure(figsize=(6,4))
sns.scatterplot(
    data=iris,
    x='sepal_length',
    y='sepal_width',
    hue='species'
)
plt.title('Scatter Plot of Sepal Length vs Sepal Width')
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(iris['sepal_length'], bins=28)
plt.title('Histogram of distribution of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(
    data=iris,
    x='species',
    y='petal_length'
)
plt.title('BoxPlot of petal Length by Species')
plt.show()
```

```python
sns.countplot(
    data=iris,
    x='species'
)
plt.title('Number of Each Species')
plt.show()
```
